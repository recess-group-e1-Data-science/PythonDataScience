# CORRELATIONS AND PLOTTING

```python
# Correlation
import pandas as pd
data = pd.read_csv('SampleData.csv')
data.corr()
```

```python
# Scatter graph
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('SampleData.csv')
data.plot(kind = 'scatter', x ='Duration',  y = 'Calories', color = 'b')
plt.title('Relationship between Duration and Calories', color = 'r', fontweight = 'bold')
plt.show()
```

```python
# Scatter graph
import pandas as pd
from matplotlib import pyplot as plt
data = pd.read_csv('SampleData.csv')
data.plot(kind = 'scatter', x ='Duration',  y = 'Pulse', color ='r')
plt.title('Relationship between Duration and Pulse', color = 'y', fontweight = 'bold')
plt.show
```

```python
# A Histogram
import pandas as pd
from matplotlib import pyplot as plt
data = pd.read_csv('SampleData.csv')
plt.hist(data.Calories)
plt.show()
```

```python
data = pd.read_csv('SampleData.csv')
data['Duration'].plot(kind = 'hist', color = 'b')
plt.xlabel('Duration')
plt.title('A HISTOGRAM FOR DURATION', color = 'purple')
plt.show()
```
