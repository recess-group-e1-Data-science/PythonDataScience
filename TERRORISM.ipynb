```python
#importing the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#setting the maximum rows and columns to display
pd.set_option('display.max_rows',127)
pd.set_option('display.max_columns',38)
```

```python
#reading the csv file into a dataframe
all_data = pd.read_csv('2018CrimeStatisticsByCategoryOfCrime.csv',header=None)
```

```python
#displaying the dataframe
all_data
```

```python
# creating the list of columns names
column_names=[
    #for reported cases
    ('REPORTED CASES','Total Cases'),('REPORTED CASES','Cases under Inquiry'),
    ('REPORTED CASES','Not Detected/NPW'),('REPORTED CASES','Cases taken to court'),
    #for performance
    ('PERFORMANCE','Cases submitted toDPP/RSA'),('PERFORMANCE','Cases with convictions'),
    ('PERFORMANCE','Cases with Acquittals'),('PERFORMANCE','Cases Dismissed'),
    ('PERFORMANCE','Cases pending in Court'),
    #for persons taken to court
    ('PERSONS TAKEN TO COURT','ACCUSED CHARGED','Male Adult'),('PERSONS TAKEN TO COURT','ACCUSED CHARGED','Male Juven.')
    ,('PERSONS TAKEN TO COURT','ACCUSED CHARGED','Female Adult'),('PERSONS TAKEN TO COURT','ACCUSED CHARGED','Female Juven.'),
    ('PERSONS TAKEN TO COURT','ACCUSED CONVICTED','Male Adult'),('PERSONS TAKEN TO COURT','ACCUSED CONVICTED','Male Juven.'),
    ('PERSONS TAKEN TO COURT','ACCUSED CONVICTED','Female Adult'), ('PERSONS TAKEN TO COURT','ACCUSED CONVICTED','Female Juven.'),
    ('PERSONS TAKEN TO COURT','ACCUSED CONVICTED','Acquitted'),('PERSONS TAKEN TO COURT','ACCUSED CONVICTED','Discharged'),
    ('PERSONS TAKEN TO COURT','ACCUSED CONVICTED','Awaiting Trial'),
    #for sentences imposed by courts
    ('SENTENCES IMPOSED BY COURTS','IMPRISONMENT','Male Adult'),
    ('SENTENCES IMPOSED BY COURTS','IMPRISONMENT','Male Juven.')
    ,('SENTENCES IMPOSED BY COURTS','IMPRISONMENT','Female Adult')
    ,('SENTENCES IMPOSED BY COURTS','IMPRISONMENT','Female Juven.'),
    ('SENTENCES IMPOSED BY COURTS','IMPRISONMENT','Death'),
    ('SENTENCES IMPOSED BY COURTS','IMPRISONMENT','Fined'),
    ('SENTENCES IMPOSED BY COURTS','IMPRISONMENT','Probation Bound Over'),
    ('SENTENCES IMPOSED BY COURTS','IMPRISONMENT','CAUTION'),
    ('SENTENCES IMPOSED BY COURTS','IMPRISONMENT','COMM. WORK'),
    ('SENTENCES IMPOSED BY COURTS','IMPRISONMENT','Fines Awarded (UgShs.)'),
    ('SENTENCES IMPOSED BY COURTS','VICTIMS OF CRIME','Male Adult'),
    ('SENTENCES IMPOSED BY COURTS','VICTIMS OF CRIME','Male Juven.')
    ,('SENTENCES IMPOSED BY COURTS','VICTIMS OF CRIME','Female Adult'),
    ('SENTENCES IMPOSED BY COURTS','VICTIMS OF CRIME','Female Juven.'),
    ('SENTENCES IMPOSED BY COURTS','VALUE PROPERTY(UGSHS.)','Lost'),
    ('SENTENCES IMPOSED BY COURTS','VALUE PROPERTY(UGSHS.)','Recovered')
]
```

```python
#selecting some from the 4th index row to bottom and 2nd index column to the last one
some_data=all_data.loc[4:,2:]
#displaying the selected columns and rows
some_data
```

```python
# setting the column names as multiIndex 
some_data.columns=pd.MultiIndex.from_tuples(column_names)
#displaying the dataframe
some_data
```

```python
# setting the index 
some_data.index=[all_data[0][4:],all_data[1][4:]]
# setting the index names 
some_data.index.names=['S/No.','CASE/CRIME INCIDENCES']
# filling all the Nan values in the column with white space 
some_data.columns= pd.MultiIndex.from_frame(some_data.columns.to_frame().fillna(''))
```

```python
#displaying the dataframe
some_data
```

```python
# filling all the nan in the index with white space 
some_data.index= pd.MultiIndex.from_frame(some_data.index.to_frame().fillna(''))
# filling the nan in the rest with - to indicate either no value or entering into a different catergory
some_data.fillna('-',inplace=True)
#displaying the dataframe
some_data
```

```python
# lets first replace all the unused cells with 0
some_data.replace('-',0,inplace=True)
# replacing the commas so as to enable us work with integers
some_data.replace(',','', regex=True, inplace=True)
temp = some_data.select_dtypes(object).columns
some_data[temp] = some_data[temp].apply(pd.to_numeric,errors='coerce')
some_data.abs()
```

```python
# filtering the rows that contains only a single value zero and replace it with the original 
some_data[~some_data.any(axis = 1)]='-'
some_data.replace('-','',inplace=True)
#displaying the dataframe
some_data
```

```python
#getting only the terrorism cases and their respective total cases
terrorism_reported_cases = some_data[('REPORTED CASES','Total Cases')].iloc[95:97]
#displaying the dataframe
terrorism_reported_cases
```

```python
#describing the dataframe using the describe() method
terrorism_reported_cases.describe()
```

```python
#Creating dataset
labels = list(terrorism_reported_cases.keys())
values = terrorism_reported_cases.values.tolist()

#Creating plot
fig1, ax1 = plt.subplots(figsize=(10,5))
ax1.pie(values, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
patches, texts, auto = ax1.pie(values, shadow=True, startangle=90, autopct='%1.1f%%')

#add justing the position of the legend on the pie char
plt.legend(patches, labels, bbox_to_anchor=(1.5,0.5), loc="lower right",bbox_transform=plt.gcf().transFigure)

#setting the title,its position and the fontsize
plt.title('A PIE CHART SHOWING THE DIFFERENT TYPES OF TERRORISM CASES AND THEIR PERCENTAGES', loc='left',fontsize=20)

# show plot
plt.show()
```

```python
#getting the total number of male juven accused and charged only
male_accused_charged_juven=some_data[('PERSONS TAKEN TO COURT','ACCUSED CHARGED','Male Juven.')].iloc[97:98].astype(int)
#getting the total number of male adult accused and charged only
male_accused_charged_adult=some_data[('PERSONS TAKEN TO COURT','ACCUSED CHARGED','Male Adult')].iloc[97:98].astype(int)
#adding the juven with the adults to get the total male accused charged
total_male_accused_charged = male_accused_charged_juven + male_accused_charged_adult
#getting the value from the series
total_male_accused_charged = total_male_accused_charged[0]
#printing the series
print('Total male accused charged: '+str(total_male_accused_charged))
```

```python
#getting the total number of female juven accused and charged only
female_accused_charged_juven=some_data[('PERSONS TAKEN TO COURT','ACCUSED CHARGED','Female Juven.')].iloc[97:98].astype(int)
#getting the total number of female adult accused and charged only
female_accused_charged_adult=some_data[('PERSONS TAKEN TO COURT','ACCUSED CHARGED','Female Adult')].iloc[97:98].astype(int)
#adding the juven and adult to get the total number of female charged accused
total_female_accused_charged = female_accused_charged_juven + female_accused_charged_adult
#getting the value from the series
total_female_accused_charged = total_female_accused_charged[0]
#printing the value
print('Total female accused charged: '+str(total_female_accused_charged))
```

```python
labels = ['Total Male Accused Charged','Total Female Accused Charged']
values = [total_male_accused_charged,total_female_accused_charged]

#Creating plot
fig1, ax1 = plt.subplots()
ax1.pie(values, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
patches, texts, auto = ax1.pie(values, shadow=True, startangle=90, autopct='%1.1f%%')

#add justing the position of the legend on the pie char
plt.legend(patches, labels, loc="right",bbox_transform=plt.gcf().transFigure)

#setting the title,its position and the fontsize
plt.title('A PIE CHART SHOWING THE PERCENTAGES OF THE FEMALE AND MALE ACCUSED CHARGED DUE TO TERRORISM', loc='left',fontsize=20)

# show plot
plt.show()
```
